```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
Co60=pd.read_table("./data/NaI_60Co_600s.dat",sep="\s+")
Cs137=pd.read_table("./data/NaI_137Cs_600s.dat",sep="\s+")
Na22=pd.read_table("./data/NaI_22Na_600s.dat",sep="\s+")
Fondo_600s=pd.read_table("./data/NaI_Fondo_600s.dat",sep="\s+")
Co57=pd.read_table("./data/NaI_57Co_300s.dat",sep="\s+")
Fondo_300s=pd.read_table("./data/NaI_Fondo_300s.dat",sep="\s+")
Canales=Fondo_600s.canal
Co60y=Co60.cuentas
Cs137y=Cs137.cuentas
Na22y=Na22.cuentas
Fondo6y=Fondo_600s.cuentas
Co57y=Co57.cuentas
Fondo3y=Fondo_300s.cuentas
```

```python
plt.plot(Canales,Co60y,label=r"$^{60}$Co")
plt.plot(Canales,Cs137y,label=r"$^{137}$Cs")
plt.plot(Canales,Na22y,label=r"$^{22}$Na")
plt.plot(Canales,Co57y,label=r"$^{57}$Co")
plt.plot(Canales,Fondo3y,label=r"Fondo 300s")
plt.plot(Canales,Fondo6y,label=r"Fondo 600s")
plt.xlim(0,500)
plt.vlines(413,0, Na22y[413],linestyles="dashed", color='black')
plt.text(420, -400, '1274 keV')
plt.vlines(228,0, Cs137y[228],linestyles="dashed", color='black')
plt.text(228, -400, '662 keV')
plt.legend()
plt.xlabel("Canales",family="serif")
plt.ylabel("Cuentas",family="serif")
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,4))
ax[0].plot(Canales,Co57y,label=r"$^{57}$Co")
ax[0].plot(Canales,Fondo3y,label=r"Fondo 300s",color="r")

ax[0].legend()
ax[0].set_xlim(0,500)
ax[0].set_xlabel("Canales",family="serif")
ax[0].set_ylabel("Cuentas",family="serif")

ax[1].plot(Canales,Co57y-Fondo3y,label=r"$^{57}$Co sin fondo")
ax[1].legend()

ax[1].set_xlim(0,500)
ax[1].set_xlabel("Canales",family="serif")
ax[1].set_ylabel("Cuentas",family="serif")
```

```python
fig,ax=plt.subplots(1,2,figsize=(11,4))
ax[0].plot(Canales,Co60y,label=r"$^{60}$Co")
ax[0].plot(Canales,Cs137y,label=r"$^{137}$Cs")
ax[0].plot(Canales,Na22y,label=r"$^{22}$Na")
ax[0].plot(Canales,Fondo6y,label=r"Fondo 600s",color="r")

ax[0].legend()
ax[0].set_xlim(0,500)
ax[0].set_xlabel("Canales",family="serif")
ax[0].set_ylabel("Cuentas",family="serif")
ax[0].set_xlim(350,600)
ax[0].set_ylim(0,1000)

ax[1].plot(Canales,Co60y-Fondo6y,label=r"$^{57}$Co sin fondo")
ax[1].plot(Canales,Cs137y-Fondo6y,label=r"$^{137}$Cs sin fondo")
ax[1].plot(Canales,Na22y-Fondo6y,label=r"$^{22}$Na sin fondo")

ax[1].legend()

ax[1].set_xlim(0,500)
ax[1].set_xlabel("Canales",family="serif")
ax[1].set_ylabel("Cuentas",family="serif")
ax[1].set_xlim(350,600)
ax[1].set_ylim(0,1000)
```
