## Punto 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,6,6)
P_cubo=np.ones(6)*1/6
P_piram=np.array([1/6,1/4,1/6,1/6,1/12,1/6])
```

Valor medio

```python
cubo_mean=np.dot(x,P_cubo)
piram_mean=np.dot(x,P_piram)
print(f"El valor medio para el dado cúbico es {cubo_mean}")
print(f"El valor medio para el dado piramidal es {piram_mean:.4f}")
```

Desviación estándar

```python
sigma_cubo=np.sqrt(np.dot((x-cubo_mean)**2,P_cubo))
sigma_piram=np.sqrt(np.dot((x-piram_mean)**2,P_piram))
print(f"La desviación estándar para el dado cúbico es {sigma_cubo:.4f}")
print(f"La desviación estándar para el dado piramidal es {sigma_piram:.4f}")
```

Asimetría

```python
asim_cubo=1/(sigma_cubo**3 )*np.dot((x-cubo_mean)**3,P_cubo)
asim_piram=1/(sigma_piram**3 )*np.dot((x-piram_mean)**3,P_piram)
print(f"La asimetría para el dado cúbico es {asim_cubo}")
print(f"La asimetría para el dado piramidal es {asim_piram:.4f}")
```

Gráficas

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))



axes[0].scatter(x,P_cubo,label=r"$P(x_i)$")
axes[0].set_title(r"$P(x)$ para dado cúbico")
axes[0].set_xlabel(r"$x$")
axes[0].set_ylabel(r"$P(x)$")
axes[0].vlines(cubo_mean,0,1, colors="red",label="Valor medio")
axes[0].text(3.55, 0.162, r"$\mu$", fontsize=12, color="red")
axes[0].hlines(0.170,-sigma_cubo+cubo_mean,sigma_cubo+cubo_mean, colors="purple",label="Desviación estándar")
axes[0].text(4.5, 0.1702, r"$\sigma$", fontsize=12, color="purple")
axes[0].set_ybound(0.1575,0.1750)
axes[0].legend(loc=3,fontsize=9)

axes[1].scatter(x,P_piram,label=(r"$P(x_i)$"))
axes[1].set_title(r"$P(x)$ para dado piramidal")
axes[1].set_xlabel(r"$x$")
axes[1].set_ylabel(r"$P(x)$")
axes[1].vlines(piram_mean,0,1, colors="red",label="Valor medio")
axes[1].text(3.3, 0.125, r"$\mu$", fontsize=12, color="red")
axes[1].hlines(0.20,-sigma_piram+piram_mean,sigma_piram+piram_mean, colors="purple",label="Desviación estándar")
axes[1].text(4.5, 0.202, r"$\sigma$", fontsize=12, color="purple")
axes[1].hlines(0.225,piram_mean,asim_piram+piram_mean, colors="green",label="Asimetría")
axes[1].text(3.4, 0.226, r"$\tau$", fontsize=12, color="green")
axes[1].set_ybound(0.075,0.26)
axes[1].legend(loc=3,fontsize=9)
plt.show()
```
