```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
me=511 #masa del electron en KeV
E1=np.array([10,100,511,1333])
eps=E1/me
theta = np.linspace(0,2*np.pi,1000)
```

```python
def f(e):
    return 0.5*(1+e*(1-np.cos(theta)))**-2   *  (1+np.cos(theta)**2+(e**2*(1-np.cos(theta))**2)/(e*(1-np.cos(theta))))
```

```python
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot(1,2,1, projection="polar")
ax2=fig.add_subplot(1,2,2)
i=0
colors = plt.cm.get_cmap('tab20', 8)
for e in eps:
    sec_ef_dif=f(e)
    ax.plot(theta,sec_ef_dif,label=f"{E1[i]}",color=colors(i))
    ax2.plot(theta,sec_ef_dif,color=colors(i))
    i+=1

fig.legend(title=r"$E_0 \ (\mathrm{keV})$")

ax.set_xticks(np.arange(0,2*np.pi,np.pi/6))
ax.set_xticklabels(["0°","30°","60°","90°","120°","150°","180°","","","","",""])
ax.set_title("Gráfica polar")
ax.set_xlabel(r"$\frac{\mathrm{d\sigma} }{\mathrm{d} \Omega} \ (r_e^2)$",fontsize=13)
ax.set_rlabel_position(-90)

ax2.set_title("Gráfica cartesiana")
ax2.grid()
ax2.set_xlim(0,np.pi)
ax2.set_xticks(np.arange(0,np.pi+0.1,np.pi/6))
ax2.set_xticklabels(["0°","30°","60°","90°","120°","150°","180°"])
ax2.set_ylabel(r"$\frac{\mathrm{d\sigma} }{\mathrm{d} \Omega} \ (r_e^2)$",fontsize=13)
ax2.set_xlabel(r"$\theta \ (°)$",fontsize=13)
plt.show()
```
